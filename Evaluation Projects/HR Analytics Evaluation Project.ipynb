## Removing outliers

zscore Method

```python
#Splitting the data to remove outliers
x= df[['Age', 'DailyRate', 'DistanceFromHome', 'Education', 'EnvironmentSatisfaction', 'HourlyRate', 'JobInvolvement', 'JobLevel', 'JobSatisfaction', 'MonthlyIncome', 'MonthlyRate', 'NumCompaniesWorked', 'PercentSalaryHike', 'PerformanceRating', 'RelationshipSatisfaction', 'StockOptionLevel', 'TotalWorkingYears', 'TrainingTimesLastYear', 'WorkLifeBalance', 'YearsAtCompany', 'YearsInCurrentRole', 'YearsSinceLastPromotion', 'YearsWithCurrManager']]
print(x.shape)
```

```python
from scipy.stats import zscore
z=np.abs(zscore(x))
z
```

```python
threshold=3
print(np.where(z>3))
```

```python
z.iloc[28][22]
```

```python
df1=df[(z<3).all(axis=1)]
print(df.shape)
print(df1.shape)
```

```python
# Data loss percentage using zscore method
a=((1470-1387)/1470)*100
a
```

IQR Method

```python
Q1 = x.quantile(0.25)
Q3 = x.quantile(0.75)
IQR=Q3-Q1
df2=x[~((x < (Q1 - 1.5 * IQR)) | (x > (Q3 + 1.5 * IQR))).any(axis=1)]
df2.shape
```

```python
# Data loss percentage using IQR method
b=((1470-779)/1470)*100
b
```

The Data loss percentage is 5.64% using zscore and 47.00% using IQR. Only 10% of data loss is acceptable. Hence i am selecting zscore method.

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
def vif_calc():
    vif=pd.DataFrame()
    vif["VIF Factor"]=[variance_inflation_factor(x.values,i) for i in range(x.shape[1])]
    vif["faetures"]=x.columns
    print(vif)
```

# Model

Finding the best random state

```python
#importing necessary libraries
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
maxAccu=0
maxRS=0
for i in range(1,200):
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.20,random_state=i)
    RF = RandomForestClassifier()
    RF.fit(x_train,y_train)
    pred=RF.predict(x_test)
    acc=accuracy_score(y_test,pred)
    if acc>maxAccu:
        maxAccu = acc
        maxRS = i
print("Best Accuracy is ",maxAccu," At random state ",maxRS)
```
