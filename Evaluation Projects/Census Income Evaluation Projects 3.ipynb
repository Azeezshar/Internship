#### Removing outliers

zscore Method

```python
#Splitting the data to remove outliers
x= df[['Age', 'Fnlwgt', 'Education_num', 'Hours_per_week']]
print(x.shape)
```

```python
from scipy.stats import zscore
z=np.abs(zscore(x))
z
```

```python
threshold=3
print(np.where(z>3))
```

```python
z.iloc[9][3]
```

```python
df1=df[(z<3).all(axis=1)]
print(df.shape)
print(df1.shape)
```

```python
# Data loss percentage using zscore method
a=((32560-31461)/32560)*100
a
```

IQR Method

```python
Q1 = x.quantile(0.25)
Q3 = x.quantile(0.75)
IQR=Q3-Q1
df2=x[~((x < (Q1 - 1.5 * IQR)) | (x > (Q3 + 1.5 * IQR))).any(axis=1)]
df2.shape
```

```python
# Data loss percentage using IQR method
b=((32560-21950)/32560)*100
b
```

The Data loss percentage is 3.37% using zscore and 32.58% using IQR. Only 10% of data loss is acceptable. Hence i am selecting zscore method.

# Model

Finding the best random state

```python
#importing necessary libraries
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
maxAccu=0
maxRS=0
for i in range(1,200):
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.20,random_state=i)
    RF = RandomForestClassifier()
    RF.fit(x_train,y_train)
    pred=RF.predict(x_test)
    acc=accuracy_score(y_test,pred)
    if acc>maxAccu:
        maxAccu = acc
        maxRS = i
print("Best Accuracy is ",maxAccu," At random state ",maxRS)
```
